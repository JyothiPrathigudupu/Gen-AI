```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

# Example: Reading the file in chunks
chunksize = 100  # Adjust based on memory constraints
file_path = 'C:/Users/nr802/Downloads/GenAI Projects/aws_review_sofware_dataset.csv'

# Initialize TF-IDF Vectorizer and Label Encoder
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
label_encoder = LabelEncoder()

# Process each chunk
processed_chunks = []
for chunk in pd.read_csv(file_path, chunksize=chunksize):
    # Convert reviewText to TF-IDF matrix
    tfidf_matrix = tfidf_vectorizer.fit_transform(chunk['reviewText'].astype(str))
    
    # Encode the 'verified' column
    chunk['verified_encoded'] = label_encoder.fit_transform(chunk['verified'])
    
    # Collect processed chunk
    processed_chunks.append(chunk)

# Combine processed chunks if needed
processed_data = pd.concat(processed_chunks, ignore_index=True)

# Print information
print("TF-IDF Matrix Shape (for one chunk):", tfidf_matrix.shape)
print(processed_data[['verified', 'verified_encoded']].head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Load the dataset
df = pd.read_csv("C:/Users/nr802/Downloads/GenAI Projects/aws_review_sofware_dataset.csv",nrows=1000)
```

```python
df.columns
```

```python
# Ensure 'text' and 'label' columns exist (modify based on actual column names)
# Assuming 'text' column contains reviews and 'label' contains sentiment
# Split data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(df['reviewText'], df['summary'], test_size=0.2, random_state=42)
df = df.dropna(subset=['reviewText', 'summary'])
```

```python

# Text preprocessing and vectorization
vectorizer = CountVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Evaluate the model
y_pred = model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.utils import class_weight

# Assuming 'reviewText' is the column with reviews and 'verified' is the sentiment label (or your target column)
# Preprocessing: Text vectorization (TF-IDF)
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
X = vectorizer.fit_transform(df['reviewText'])

# Split data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, df['verified'], test_size=0.2, random_state=42)  # Correct target column

# Compute class weights for imbalanced data
class_weights = class_weight.compute_class_weight('balanced', classes=df['verified'].unique(), y=df['verified'])  # Corrected target column
class_weight_dict = dict(zip(df['verified'].unique(), class_weights))  # Creating the class weight dictionary

# Train a Logistic Regression model with class weights
model = LogisticRegression(class_weight=class_weight_dict)  # Corrected syntax
model.fit(X_train, y_train)

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
