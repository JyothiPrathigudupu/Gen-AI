```python
from langchain.llms import OpenAI

# Initialize the model with streaming enabled, if required
model = OpenAI(streaming=True)

# Define the prompt
messages = [
    {"role": "user", "content": "assume you are a chief in a hotel make chicken biryani step by step process until it cooked well "}
]

# Stream the response and print each part as it arrives
response_stream = model.stream(messages)
for message in response_stream:
    print(message, end="", flush=True)  # Print message directly if it's a string
```
